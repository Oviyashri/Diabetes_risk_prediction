# Early Stage Diabetes Risk Prediction Dataset

```python
import numpy as np
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import LinearSVC,SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
```

```python
df=pd.read_csv('D:/Python/Early stage diabetes risk prediction dataset/diabetes_data_upload.csv')
df
```

```python
df.info()
```

```python
print("Number of null values:", df.isnull().sum().sum())
```

```python
df.describe(include='all')
```

```python
{column:len(df[column].unique())
for column in df.columns}
```

```python
df["Gender"]=df["Gender"].replace({'Male':1,'Female':0})
print(df["Gender"])
```

```python
for column in df.columns.drop(['Age','Gender','class']):
 df[column]= df[column].replace({'No':0, 'Yes':1})
print(df)
```

```python
y=df["class"]
X=df.drop("class",axis=1)
X
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.7,shuffle=True,random_state=46)
```

```python
X_train
```

```python
y_test
```

```python
model1=LogisticRegression()
model1.fit(X_train,y_train)
model1.score(X_test,y_test)
model1.score(X_test,y_test)*100
```

```python
model2=KNeighborsClassifier()
model2.fit(X_train,y_train)
model2.score(X_test,y_test)
model2.score(X_test,y_test)*100
```

```python
model3=LinearSVC()
model3.fit(X_train,y_train)
model3.score(X_test,y_test)
model3.score(X_test,y_test)*100
```

```python
model4=SVC()
model4.fit(X_train,y_train)
model4.score(X_test,y_test)
model4.score(X_test,y_test)*100
```

```python
model5=DecisionTreeClassifier()
model5.fit(X_train,y_train)
model5.score(X_test,y_test)
model5.score(X_test,y_test)*100
```

```python
model6=RandomForestClassifier()
model6.fit(X_train,y_train)
model6.score(X_test,y_test)
model6.score(X_test,y_test)*100
```

```python
model7=GradientBoostingClassifier()
model7.fit(X_train,y_train)
model7.score(X_test,y_test)
model7.score(X_test,y_test)*100
```

```python
print(model1.score(X_test,y_test)*100)
print(model2.score(X_test,y_test)*100)
print(model3.score(X_test,y_test)*100)
print(model4.score(X_test,y_test)*100)
print(model5.score(X_test,y_test)*100)
print(model6.score(X_test,y_test)*100)
print(model7.score(X_test,y_test)*100)
```
